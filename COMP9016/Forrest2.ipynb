```python
class FiremanAgent:
    def __init__(self):
        self.actions = ['clean', 'move_right', 'move_down']
        self.current_action = 0

    def choose_action(self, percept):
        if percept == 'fire':
            return 'fire'
        else:
            return self.actions[self.current_action]

    def update_action(self):
        self.current_action = (self.current_action + 1) % len(self.actions)
```

```python
from random import choice

########################################
# ENVIRONMENT CLASS
########################################
class Environment2D(GraphicEnvironment):
    def __init__(self):
        # Initialize a 2x2 grid with random dirt distribution
        self.grid = [[random.choice(['empty', 'fire']) for _ in range(2)] for _ in range(2)]
        self.agent_location = (0, 0)

    def is_fire(self, x, y):
        return self.grid[x][y] == 'fire'

    def empty(self, x, y):
        self.grid[x][y] = 'empty'

    def percept(self):
        x, y = self.agent_location
        return self.is_fire(x, y)

    def step(self, action):
        x, y = self.agent_location
        if action == 'fire':
            self.fire(x, y)
        elif action == 'move_right':
            if y < 1:
                self.agent_location = (x, y + 1)
        elif action == 'move_down':
            if x < 1:
                self.agent_location = (x + 1, y)



# Forrest Class
class Forrest2D(Environment):
    # def percept(self, agent):
    #     '''return a list of things that are in our agent's location'''
    #     things = self.list_things_at(agent.location)
    #     return things
    def percept(self, agent):
        """By default, agent perceives things within a default radius."""
        return agent.location, self.list_things_at(agent.location)
    
    def thing_classes(self):
        return [Fire, Water, ModelFireman]

    def execute_action(self, agent, action):
        if action == 'TurnRight':
            agent.direction += Direction.R
        elif action == 'TurnLeft':
            agent.direction += Direction.L
        elif action == 'Forward':
            agent.bump = self.move_to(agent, agent.direction.move_forward(agent.location))
            
        if action != 'NoOp':
            agent.performance -= 1
            

    def is_done(self):

        return False
```
